```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

print("Libraries imported successfully.")
```

```python
df = pd.read_csv(
    "../data/raw/MachineLearningRating_v3.txt",
    delimiter="|"
)

df.columns = df.columns.str.strip()

print(df.shape)

df.head()
```

```python
df["Province"].value_counts().head(10)
```

```python
df["HasClaim"] = (df["TotalClaims"] > 0).astype(int)

df["Margin"] = df["TotalPremium"] - df["TotalClaims"]

df["LossRatio"] = np.where(
    df["TotalPremium"] > 0,
    df["TotalClaims"] / df["TotalPremium"],
    np.nan
)

print("Features created successfully.")
```

```python
group_a = df[df["Province"] == "Gauteng"]["HasClaim"]

group_b = df[df["Province"] == "Western Cape"]["HasClaim"]
```

```python
t_stat, p_value = stats.ttest_ind(
    group_a,
    group_b,
    equal_var=False,
    nan_policy="omit"
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis.")
    print("There is a statistically significant difference in risk.")
else:
    print("Fail to reject the null hypothesis.")
    print("No statistically significant difference detected.")
```

```python
df["PostalCode"].value_counts().head(10)
```

```python
zip_a = df[df["PostalCode"] == 2000]["HasClaim"]

zip_b = df[df["PostalCode"] == 8001]["HasClaim"]
```

```python
t_stat, p_value = stats.ttest_ind(
    zip_a,
    zip_b,
    equal_var=False,
    nan_policy="omit"
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis.")
    print("Risk differs significantly between postal codes.")
else:
    print("Fail to reject the null hypothesis.")
    print("No significant risk difference detected.")
```

```python
margin_a = df[df["PostalCode"] == 2000]["Margin"]

margin_b = df[df["PostalCode"] == 8001]["Margin"]
```

```python
t_stat, p_value = stats.ttest_ind(
    margin_a,
    margin_b,
    equal_var=False,
    nan_policy="omit"
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis.")
    print("Margins differ significantly between postal codes.")
else:
    print("Fail to reject the null hypothesis.")
    print("No significant margin difference detected.")
```

```python
men = df[df["Gender"] == "Male"]["HasClaim"]

women = df[df["Gender"] == "Female"]["HasClaim"]
```

```python
t_stat, p_value = stats.ttest_ind(
    men,
    women,
    equal_var=False,
    nan_policy="omit"
)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Reject the null hypothesis.")
    print("There is a significant gender risk difference.")
else:
    print("Fail to reject the null hypothesis.")
    print("No significant gender risk difference detected.")
```
