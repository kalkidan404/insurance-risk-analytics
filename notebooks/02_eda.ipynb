```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

print("Libraries imported successfully.")
```

```python
df = pd.read_csv(
    "../data/raw/MachineLearningRating_v3.txt",
    delimiter="|"
)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["HasClaim"] = (df["TotalClaims"] > 0).astype(int)
```

```python
df["HasClaim"].mean()
```

```python
claims = df[df["TotalClaims"] > 0]
```

```python
claims["TotalClaims"].describe()
```

```python
df.groupby("Province")[["TotalClaims", "TotalPremium"]].sum()
```

```python
province_lr = df.groupby("Province")[["TotalClaims", "TotalPremium"]].sum()

province_lr["LossRatio"] = province_lr["TotalClaims"] / province_lr["TotalPremium"]
```

```python
df["LossRatio"] = np.where(
    df["TotalPremium"] > 0,
    df["TotalClaims"] / df["TotalPremium"],
    np.nan
)

df["Margin"] = df["TotalPremium"] - df["TotalClaims"]
```

```python
missing_values = df.isnull().sum()

missing_values[missing_values > 0]
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df["TotalClaims"], bins=50)

plt.title("Distribution of Total Claims")

plt.show()
```

```python
duplicate_rows = df.duplicated().sum()

print(f"Duplicate Rows: {duplicate_rows}")
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_percent = missing_percent.sort_values(ascending=False)

missing_percent[missing_percent > 0]
```

```python
numerical_cols = df.select_dtypes(include=["int64", "float64"]).columns

categorical_cols = df.select_dtypes(include=["object"]).columns

print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)
```

```python
overall_loss_ratio = df["TotalClaims"].sum() / df["TotalPremium"].sum()

print(f"Overall Loss Ratio: {overall_loss_ratio:.2f}")
```

```python
province_loss_ratio = (
    df.groupby("Province")[["TotalClaims", "TotalPremium"]]
    .sum()
)

province_loss_ratio["LossRatio"] = (
    province_loss_ratio["TotalClaims"] /
    province_loss_ratio["TotalPremium"]
)

province_loss_ratio = province_loss_ratio.sort_values(
    by="LossRatio",
    ascending=False
)

province_loss_ratio
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x=province_loss_ratio.index,
    y=province_loss_ratio["LossRatio"]
)

plt.title("Loss Ratio by Province")

plt.xticks(rotation=45)

plt.ylabel("Loss Ratio")

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(x=df["TotalClaims"])

plt.title("Boxplot of Total Claims")

plt.show()
```

```python
sample_df = df.sample(5000, random_state=42)

plt.figure(figsize=(10,6))

sns.scatterplot(
    data=sample_df,
    x="TotalPremium",
    y="TotalClaims",
    alpha=0.5
)

plt.title("Total Premium vs Total Claims")

plt.show()
```

```python
selected_cols = [
    "TotalPremium",
    "TotalClaims",
    "CustomValueEstimate",
    "CalculatedPremiumPerTerm",
    "LossRatio",
    "Margin"
]

correlation_matrix = df[selected_cols].corr()

correlation_matrix
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Matrix")

plt.show()
```

```python
df["TransactionMonth"] = pd.to_datetime(
    df["TransactionMonth"]
)
```

```python
monthly_claims = (
    df.groupby("TransactionMonth")["TotalClaims"]
    .sum()
)

monthly_claims.head()
```

```python
plt.figure(figsize=(12,6))

monthly_claims.plot()

plt.title("Monthly Total Claims Trend")

plt.ylabel("Total Claims")

plt.show()
```

```python
vehicle_risk = df.groupby("VehicleType")[["TotalClaims", "TotalPremium"]].sum()

vehicle_risk["LossRatio"] = vehicle_risk["TotalClaims"] / vehicle_risk["TotalPremium"]

vehicle_risk.sort_values("LossRatio", ascending=False)
```

```python
gender_risk = df.groupby("Gender")[["TotalClaims", "TotalPremium"]].sum()

gender_risk["LossRatio"] = gender_risk["TotalClaims"] / gender_risk["TotalPremium"]

gender_risk
```
